```python
from sklearn.metrics import precision_recall_curve
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np

# Ruta a la carpeta con los archivos de etiquetas
labels_folder = '/home/mosquitos/Documentos/TesisDiegoMathi/Mosquitos-1/valid/labels'

# Diccionario para almacenar las etiquetas verdaderas por imagen
true_labels_dict = {}

# Recorrer todos los archivos de etiquetas
for label_file in os.listdir(labels_folder):
    if label_file.endswith('.txt'):
        image_id = label_file.replace('.txt', '')  # Obtener el nombre base del archivo
        true_labels_dict[image_id] = []
        
        with open(os.path.join(labels_folder, label_file), 'r') as file:
            for line in file:
                class_id = int(line.split()[0])  # Extraer la clase
                true_labels_dict[image_id].append(class_id)

# Cargar el modelo entrenado
model = YOLO('/home/mosquitos/Documentos/TesisDiegoMathi/runs/detect/epoch100_yolov8l8/weights/best.pt')

# Realizar inferencia en el conjunto de im√°genes
results = model.predict('/home/mosquitos/Documentos/TesisDiegoMathi/Mosquitos-1/valid/images', save=True, conf=0.5)

# Listas para almacenar puntuaciones de predicci√≥n y etiquetas verdaderas por clase
predicted_scores_0, true_labels_0 = [], []
predicted_scores_1, true_labels_1 = [], []

# Extraer puntuaciones de confianza de las predicciones y etiquetas verdaderas
for result in results:
    image_id = os.path.basename(result.path).split('.')[0]  # Obtener el ID de la imagen sin extensi√≥n
    true_classes = true_labels_dict.get(image_id, [])  # Clases reales de la imagen
    
    if result.boxes is not None:  # Verificar si hay detecciones
        for box in result.boxes:
            class_pred = int(box.cls.item())  # Clase predicha
            score = box.conf.item()  # Confianza

            if class_pred == 0:  # Clase 0
                true_label = 1 if 0 in true_classes else 0
                predicted_scores_0.append(score)
                true_labels_0.append(true_label)
                print(f"[Clase 0] Imagen: {image_id}, Predicci√≥n: {class_pred}, True Label: {true_label}")

            elif class_pred == 1:  # Clase 1
                true_label = 1 if 1 in true_classes else 0
                predicted_scores_1.append(score)
                true_labels_1.append(true_label)
                print(f"[Clase 1] Imagen: {image_id}, Predicci√≥n: {class_pred}, True Label: {true_label}")

# Convertir listas a NumPy arrays para evitar errores
true_labels_0 = np.array(true_labels_0)
predicted_scores_0 = np.array(predicted_scores_0)
true_labels_1 = np.array(true_labels_1)
predicted_scores_1 = np.array(predicted_scores_1)

# Verificar la cantidad de etiquetas
print(f"Total etiquetas clase 0: {len(true_labels_0)}, Total predicciones clase 0: {len(predicted_scores_0)}")
print(f"Total etiquetas clase 1: {len(true_labels_1)}, Total predicciones clase 1: {len(predicted_scores_1)}")

# Asegurar que haya datos antes de calcular la curva
if len(true_labels_0) > 0 and len(predicted_scores_0) > 0:
    precision_0, recall_0, _ = precision_recall_curve(true_labels_0, predicted_scores_0)
else:
    precision_0, recall_0 = [0], [0]  # Valores por defecto si no hay datos

if len(true_labels_1) > 0 and len(predicted_scores_1) > 0:
    precision_1, recall_1, _ = precision_recall_curve(true_labels_1, predicted_scores_1)
else:
    precision_1, recall_1 = [0], [0]  # Valores por defecto si no hay datos

# Graficar las curvas
plt.figure(figsize=(8, 6))
plt.plot(recall_0, precision_0, marker='.', label='Clase 0')
plt.plot(recall_1, precision_1, marker='.', label='Clase 1')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
import optuna
from ultralytics import YOLO
```

```python
def objective(trial):
    lr0 = trial.suggest_loguniform('lr0', 1e-5, 1e-2)
    momentum = trial.suggest_uniform('momentum', 0.6, 0.98)
    epochs = trial.suggest_int('epochs', 10, 100)
    imgsz = trial.suggest_categorical('imgsz', [320, 640, 1280])

    model = YOLO("yolov8l.pt")

    results = model.train(
        data="/home/mosquitos/Documentos/TesisDiegoMathi/Mosquitos-1/data.yaml",
        epochs=epochs,
        imgsz=imgsz,
        lr0=lr0,
        momentum=momentum,
        batch=8,
        amp=True,
        name=f"trial_{trial.number}"
    )
    
    val_map = results.results_dict.get('metrics/mAP50(B)', None)
    if val_map is None:
        raise ValueError("La m√©trica 'metrics/mAP50(B)' no est√° disponible.")

    return val_map
```

```python
optuna.logging.set_verbosity(optuna.logging.INFO)
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10)

# Imprimir los mejores hiperpar√°metros
print("Mejores hiperpar√°metros:", study.best_params)
```

```python
best_params = study.best_params

model = YOLO("yolov8l.pt")
results = model.train(
    data="/home/mosquitos/Documentos/datasets/Clasificador-de-mosquitos-3/data.yaml",
    epochs=150,
    imgsz=best_params['imgsz'],
    lr0=best_params['lr0'],
    momentum=best_params['momentum'],
    batch=8,
    amp=True,
    name="final_model"
)
```
